# np.corrcoef

````python
"""
it drescribe how the strengh of the relationship between variables
it returns the value between -1 to 1 i means to variables are strongly postive related,
-1 means strongly negative related , 
0 means no relationship between them
"""
````

````python
a = np.arange(1,100,10)                #    a , b
b = np.arange(1,100,10)              # a    1   1
np.corrcoef(a,b)                     # b    1   1
````

````python
a = np.random.randint(1,100,10)
b = np.random.randint(1,100,10)   
np.corrcoef(a,b)
````
